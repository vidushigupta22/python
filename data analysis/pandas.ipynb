# Difference b/w series and dataframes

***SERIES -> its a one dimensional array which stores same type of data like a particular column or row. One can say comination of series results in DataFrame***

***DATAFRAME --> its a tabular representation of data , where rows and columns are reflected***

```python
# pandas work with data frames i.e data displayed in spreadsheet manner 
```

```python
import pandas as pd
import numpy as np
```

```python
# lets make dataframe using pandas
dict1 = {
    'Name' : ['Vidushi','Ayukti','Khushi','Ashi'],
    'Marks' : [78,95,76,83],
    'Age' : [18,18,17,19],
    'class' : [12,11,10,11]
}
```

```python
# to convert dictionary into dataframe

df = pd.DataFrame(dict1)
df

# now 0,1,2,3 are indexes which refer to rows and columns = name marks age class
```

```python
# one can convert this data frame into csv file so that it can be open in excel

df.to_csv('class.csv')

# if you open this class file then index will be displayed in first column of excel sheet, to prevent that 
df.to_csv('class.csv',index = False)
```

```python
# to view first 2 rows

df.head(2)
```

```python
# to view last 2 rows

df.tail(2)
```

```python
# to get the statisticl analysis of numerical columns in data dataframe
df.describe()
```

```python
# one can also read the csv files in python via pandas and make changes too

pd.read_csv('class.csv')
```

```python
# one can also change the values
df['Name'][3] = 'Meghna'
```

```python
# again convert it and see the change
df.to_csv('class.csv',index = False)
pd.read_csv('class.csv')
```

# Cleaning Data Sets

***dropna()  --> function is used to remove rows and columns with Null/NaN values***

***By default, this function returns a new DataFrame and the source DataFrame remains unchanged***

```python
car_sales = pd.DataFrame({'Car':['audi','bmw','porche','mercedes'],
'Place' : ['kolkata','pune','jaipur','mumbai'],
'UnitsSold' : [20,np.nan,40,np.nan]})
```

```python
car_sales
```

```python
# to drop null values

car_sales.dropna()
```

```python
# if a column contains all null values then use --> car_sales.dropna(how = all)
```
