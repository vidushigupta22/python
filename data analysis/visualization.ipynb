```python
x = np.arange(-10,10,0.1)
x2 = x ** 2
x3 = x**3 
```

```python
plt.figure(figsize=(15,7))
plt.plot(x,label = 'original')
plt.plot(x2,label = 'squared')
plt.plot(x3,label = 'cube')
plt.legend() # activate the legend 
plt.xlabel = ('x axis')
plt.ylabel = ('y axis')
plt.title = ('visualization of x,squares and cubes')
plt.show()
```

```python
# comparision is better visualized by bar plot and pie chart

cities = ['kolkata','chennai','pune','banglore','lucknow']
rainfall = [10,50,60,40,80]
```

```python
plt.bar(cities,rainfall)
```

```python
plt.barh(cities,rainfall) # horizontal representation
```

```python
plt.figure(figsize=(15,7))
plt.xlabel=("CITIES")
plt.ylabel = ("amount of rainfall")
plt.bar(cities,rainfall,color = 'orange',edgecolor = 'red',hatch = '//',width=(.1,.2,.3,.4,.5))
plt.show()
```

```python
plt.figure(facecolor = 'white')
wedges = [0,0,0,0,0.3]
plt.pie(rainfall,labels=cities,autopct = '%1.1f%%',radius = 2,explode = wedges)
plt.show()
```

```python
prices = [213,34,46,678,67,8,78,34,56,23]
qty = [12,45,67,89,34,23,12,45,23,78]
quality = [12,45,68,89,3,41,2,3,7,10]
```

```python
# scatter is used when there is relation between data
plt.scatter(prices,qty)
```

```python
plt.scatter(prices,qty,color = 'purple',marker = "*",s = 100) # s represents size of marker 
```

```python
plt.scatter(prices,qty,color = 'purple',marker = "H",s = 100*3)
plt.scatter(prices,quality,color = 'RED',marker = "x",alpha =.5 ,s = 100*3,edgecolors='c=BLACK')
```
